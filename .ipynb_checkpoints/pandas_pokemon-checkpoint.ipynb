```python
#export in diff formats
df.to_csv('reordered_data.csv', index=False)
df.to_excel('reordered_data.xlsx', index=False)
df.to_csv('reordered_data.txt', index=False, sep='\t')
```

### Filtering 

```python
# 1 condition
df.loc[(df['Type 1']=='Grass')]
```

```python
# 2 conditions
df.loc[(df['Type 1']=='Grass')&(df['Type 2']=='Poison')].head(10)
```

```python
# 2 conditions (or)
df.loc[(df['Type 1']=='Grass')|(df['Type 2']=='Poison')].head(10)
```

```python
# 3 conditions 
df.loc[(df['Type 1']=='Grass')&(df['Type 2']=='Poison')&(df['HP']>70)].head(10)
```

```python
new_df_sub = df.loc[(df['Type 1']=='Grass')&(df['Type 2']=='Poison')&(df['HP']>70)]
new_df_sub.head()
```

```python
new_df_sub.reset_index(drop=True, inplace=True)
```

```python
new_df_sub
```

```python
#find by text
df.loc[df['Name'].str.contains('Mega')].head()
```

```python
#does not contain
df.loc[~df['Name'].str.contains('Mega')].head()
```

Regular Expressions

```python
import re 
```

```python
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
```

```python
#ignore case
df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('pi[a-z]*', flags = re.I, regex=True)]
```

```python
#beginning with Pi
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex=True)]
```

### Group By

```python
df.groupby('Type 1').mean().sort_values('Defense',ascending=False)
```

```python
df.groupby('Type 1').mean().sort_values('HP',ascending=False)
```

```python
df['count'] = 1
df.groupby(['Type 1','Type 2']).count()['count']
```

```python
df.groupby(['Type 1']).mean()
```

```python
df.groupby('Type 1').Defense.mean().sort_values(ascending=False)
```

```python
df.groupby('Type 1').HP.max().sort_values(ascending=False)
```
